```python
grocery_list = ["Rice", "Milk", "Eggs", "Bread", "Sugar"]

print("Grocery List:")
print(grocery_list)
grocery_list.append("Oil")
grocery_list.append("Fruits")
print("\nAfter adding new items:")
print(grocery_list)
grocery_list.remove("Milk")
grocery_list.remove("Sugar")
print("\nAfter removing purchased items:")
print(grocery_list)
print("\nAccessing items using indexing:")
print("First item:", grocery_list[0])
print("Last item:", grocery_list[-1])
print("\nFinal Shopping List:")
for item in grocery_list:
    print(item)
```
